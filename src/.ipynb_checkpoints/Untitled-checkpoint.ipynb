```python
import pickle
fn = ("data/vehicle_value_data_temp"
      +".pickle")
with open(fn,'rb') as read_file:
    t_df = pickle.load(read_file)

t_df
```

```python
def get_cty_mpg_val(t):
    tt = t.split(" hwy")[0].split("cty / ")
    tt = tt[0]
    return tt
def get_hwy_mpg_val(t):
    tt = t.split(" hwy")[0].split("cty / ")
    tt = tt[-1]
    return tt
def get_engine_size(t):
    tt = t.split(" ")[0].split('l')
    tt = tt[0]
    return tt
    
t_df["cty_mpg"] = t_df["mpg"].apply(get_cty_mpg_val)
t_df["hwy_mpg"] = t_df["mpg"].apply(get_hwy_mpg_val)
t_df["cty_mpg"][t_df["cty_mpg"] != 'n/a '] =  t_df["cty_mpg"][t_df["cty_mpg"] != 'n/a '].astype(int)
t_df["cty_mpg"][t_df["cty_mpg"] == 'n/a '] = None

t_df["hwy_mpg"][t_df["hwy_mpg"] != 'n/a'] =  t_df["hwy_mpg"][t_df["hwy_mpg"] != 'n/a'].astype(int)
t_df["hwy_mpg"][t_df["hwy_mpg"] == 'n/a'] = None

t_df["year"] = t_df["year"].astype(int)
t_df["mileage"] = t_df["mileage"].str.split(",").str.join('')
t_df["fuel type"][t_df["fuel type"]=="n/a"] = None

t_df["price"] = t_df["price"].str[1:].str.split(",").str.join('')
t_df["drive type"][t_df["drive type"] == "4wd"] = "awd"
t_df["transmission"][t_df["transmission"] == "n/a"] = None

t_df["engine_size"] = t_df["engine"].apply(get_engine_size)
t_df["engine_size"][t_df["engine_size"]==""] = None
t_df
```
